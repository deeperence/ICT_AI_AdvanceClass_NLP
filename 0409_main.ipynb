```python
from sklearn.metrics import confusion_matrix

# Binomial에 대한 confusion matrix 응용
# 0 : 스팸, 1 : 정상, D : 6
trueY = [1,0,1,1,0,1]
predY = [0,0,1,1,0,1]

P:(2/3)*(3/3)
R:()
    
result = confusion_matrix(trueY, predY) # scikit learn의 confusion matrix는 transpose되어있음. 
```

(이미지 속 P와 R은 잘못 계산된 값)




* Precision : 모델이 예측한 정답 중 진짜 정답의 비율
* Recall : 진짜 정답 중 모델이 찾은 정답의 비율

```python
result = result.transpose()
result
```

```python
from sklearn.metrics import classification_report
print(classification_report(trueY, predY))
```

```python
for i in range(result.shape[0]):
    print("Precision: {0}".format(result[i, i] / result[i].sum()))
    print("Recall: {0}".format(result[i, i] / result[:, i].sum()))
```

```python
# classification_report를 직접 구현해 보기(class = 3일때 precision과 recall 구해보기)
trueY = [2, 0, 2, 2, 0, 1]
predY = [0, 0, 2, 2, 0, 2]

# class0:  P=2/3, R=2/2
# class1:   P=0/0, R=0/1
# class2:  P=2/3, R=2/3

result = confusion_matrix(trueY, predY)
result
```

```python
result = result.transpose()

for i in range(result.shape[0]):
    print('Precision: {0}'.format(result[i, i] / result[i].sum()), 'Recall: {0}'.format(result[i, i] / result[:, i].sum()))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(trueY, predY))
```
